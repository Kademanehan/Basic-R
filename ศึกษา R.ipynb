## ชนิดและรูปแบบของข้อมูลใน R

> - Numeric   -เป็นข้อมูลที่มีค่าเป็นตัวเลข เช่น อายุ รายได้ และคะแนนสอบ เป็นต้น ซึ่งข้อมูลชนิดนี้สามารถนำมาคำนวณหาค่าทางสถิติต่าง ๆ 
> - Character -เป็นข้อมูลที่ค่าของมันไม่ใช่ตัวเลข โดยปรกติมักเป็นข้อความ หรือตัวอักษร ในทางสถิติเรียกข้อมูลชนิดนี้ว่าข้อมูลเชิงคุณภาพ เช่น เพศ (ชาย, หญิง) เป็นต้น
> - Logical   -เป็นข้อมูลเชิงตรรกที่มีค่าอยู่สองค่าเท่านั้น คือ TRUE แทนค่าเป็นจริง หรือ FALSE แทนค่าที่เป็นเท็จ
> - Factor    -เป็นลักษณะการเก็บข้อมูลที่เป็นข้อมูลเชิงกลุ่ม หรือข้อมูลเชิงคุณภาพ ที่ไม่ได้มีค่าเป็นตัวเลขอย่างแท้จริง แต่ใช้ในการจำแนกข้อมูลออกเป็นกลุ่ม ๆ ตามลักษณะเฉพาะที่สนใจ
> - Vector    -ข้อมูลที่อยู่ในรูปของเวกเตอร์อาจเป็นตัวเลข หรือตัวอักษร ก็ได้ โดยสมาชิกในเวกเตอร์ต้องเป็นข้อมูลชนิดเดียวกัน ลักษณะการเก็บข้อมูลของเวกเตอร์ใน R จะเป็นเวกเตอร์แถว 
> - Matrix    -เป็นลักษณะของการเก็บข้อมูลในสองมิติ (2 dimensions) ข้อมูลที่อยู่ในรูปของเมทริกซ์อาจเป็นตัวเลข หรือตัวอักษรก็ได้ แต่สมาชิกทุกตัวในเมทริกซ์หนึ่งต้องเป็นข้อมูลชนิดเดียวกันเท่านั้น
> - Data frame -เป็นลักษณะการเก็บข้อมูลในรูปตาราง ที่ประกอบด้วยเวกเตอร์หนึ่งหรือหลายเวกเตอร์มารวมกัน แต่ความยาวของทุกเวกเตอร์ต้องเท่ากัน
> - List       -เป็นลักษณะการเก็บข้อมูลที่สามารถนำ object แต่ละแบบมารวมกันเก็บเป็น list ได้ จะเป็นข้อมูลชนิดเดียวกันหรือไม่ก็ได่

```r
library(tidyverse)
```

## การเลือกข้อมูลตามเงื่อนไข (Filter)

เราสามารถใช้คำสั่ง filter() ในการเลือกแถว (rows) จากค่าของข้อมูลในแต่ละคอลัมน์ 
เครื่องหมายที่สร้างเงื่อนไขใน R ได้มีดังนี้

>- เครื่องหมาย != หมายถึง ไม่เท่ากับ
>- เครื่องหมาย  > หมายถึง มากกว่า
>- เครื่องหมาย  < หมายถึง น้อยกว่า
>- เครื่องหมาย >= หมายถึง มากกว่าหรือเท่ากับ
>- เครื่องหมาย <= หมายถึง น้อยกว่าหรือเท่ากับ
>- เครื่องหมาย %in% หมายถึง เป็นส่วนหนึ่งของเวคเตอร์ที่ตามมา

```r
prodoae |> 
  filter(subcommod %in% c("ข้าวนาปี", "ข้าวนาปรัง"), 
         year >= 2524, year <= 2540)
# เงื่อนไข AND สามารถใช้เครื่องหมาย & และ , 
# เงื่อนไข OR สามารถใช้เครื่องหมาย |
# เลือก filter ข้อมูลข้าวนาปีและข้าวนาปรัง กำหนดแสดงข้อมูลตั้งแต่ปี 2524-2540
```

## การเรียงลำดับข้อมูล (Arrange)

คำสั่ง arrange คล้ายคลึงกับคำสั่ง sort ใน Excel เราสามารถเรียงลำดับแถวตามค่าในคอลัมน์

```r
prodoae |>
 arrange(province)
# เรียงลำดับข้อมูลตามคอลัมน์ Province
```

ถ้าเราต้องการกลับกันให้ใส่คำสั่ง desc()

```r
prodoae |>
  filter(subcommod == "ยางพารา") |>
  arrange(province, desc(year)) |> head(10)
# filter ยางพารา โดยให้เรียงตามจังหวัด และปีมากไปน้อยมาก่อน
```

## การเลือกข้อมูล (Select)

เราใช้ select() เลือกเฉพาะคอลัมน์ที่เราต้องการ

```r
prodoae |>
  select(subcommod, year, province, production)
# หรือเอาคอลัมน์ที่เราไม่ต้องการออก โดยการใส่เครื่องหมาย - ไปด้านหน้า
#/> prodoae |> select(subcommod, year, -province, -production)
```

## การสร้างตัวแปรใหม่ (Mutate)

คำสั่ง mutate นี้จะช่วยสร้างคอลัมน์ให้เรา โดยอาศัยข้อมูลที่มีอยู่แล้วในคอลัมน์เดิม หรือ เพิ่มข้อมูลใหม่เข้าไปก็ได้

```r
prodoae |>
  mutate(production_m = production/1000000) #|>
#   select(subcommod, year, province, area_plant, area_harvest, 
#          yield_plant ,yield_harvest, production, production_m) #|>
#   filter(province == "ประเทศไทย"& subcommod == "ข้าวเจ้าหอมมะลินาปี")
# การสร้างตัวแปรใหม่ ชื่อ production_m โดยคำสั่ง mutate หาร 1,000,000 เพื่อแปลงผลผลิตหน่วย ตัน เป็นหน่วย ล้านตัน 
```

## การสรุปข้อมูล (Summarize)

```r
prodoae |>
    filter(!is.na(reg_nesdb)) |>
    group_by(subcommod, year, reg_nesdb) |>
    summarize(production = mean(area_harvest, na.rm = FALSE)/1e6)
# filter ข้อมูลโดยเช็คค่าที่ไม่มีข้อมูลการแบ่งภาคก่อน 
# การ Group by ให้แสดงข้อมูลสินค้าแต่ละประเภท ในแต่ละปีของแต่ละภาค ว่ามีผลผลิตเท่าไหร่ (แปลงหน่วย ตัน เป็น ล้านตัน)
# จากนั้นทำการ Summarise ค่าเฉลี่ยของผลผลิตแต่ละประเภท และใช้คำสั่ง na.rm เพื่อไม่ให้นับจำนวนค่าที่ไม่มีข้อมูล
```

# Data Visualization

## Bar Chart

```r
ggplot(
  data = prodoae |> 
    left_join(tha1, by = c("province_code" = "adm1_pcode")) |>
    filter(year == 2564, 
           subcommod == "ยางพารา",
           reg_nesdb == "C"),
  aes(
    x = yield_harvest, 
    y = reorder(adm1_en, yield_harvest),
    fill = adm1_en)) + 
  geom_col()
```

## Line Chart

```r
ggplot(
  data = prodoae |> 
    mutate(production_m = production/1000000)|>
    filter(province %in% c("กระบี่","อุดรธานี"), 
         year >= 2540, year <= 2564,
         subcommod == "ยางพารา"),
  aes(
    x = year, 
    y = production_m, 
    color = province)) + 
  geom_line()
```

## Histogram

```r
ggplot(
  data = prodoae |>
    mutate(production_m = production/1000000)|>
    filter(year == 2560),
  aes(x = production_m)) + 
  geom_histogram(
    bins = 10, 
    fill = "navyblue", 
    color = "white")
```

```r
ggplot(data = prodoae |> 
           filter(!is.na(reg_nesdb), subcommod == "ข้าวนาปี"), 
       aes(x = year, 
           y = yield_plant, 
#            size = pop,
           col = reg_nesdb)) + 
  geom_point(alpha = 0.5) +
  scale_x_log10() +
  scale_size(range = c(1, 15))
```
